# Informe Final – Análisis de rendimiento de las tiendas Alura Store

## Introducción

El objetivo de este análisis es evaluar el desempeño de las cuatro tiendas de la cadena **Alura Store** para ayudar al Sr. Juan a tomar una decisión estratégica sobre cuál tienda debería vender para invertir en un nuevo negocio.

Para ello, se analizaron diferentes indicadores clave del desempeño comercial de cada tienda, incluyendo **los ingresos totales, las categorías de productos más y menos vendidas, las calificaciones promedio de los clientes, los productos más y menos vendidos y el costo promedio de envío**. Además, se utilizaron visualizaciones para facilitar la comparación entre las tiendas y detectar patrones relevantes en los datos.

---

## Análisis de ingresos

El primer indicador analizado fue el ingreso total generado por cada tienda:

- **Tienda 1:** 1,150,880,400  
- **Tienda 2:** 1,116,343,500  
- **Tienda 3:** 1,098,019,600  
- **Tienda 4:** 1,038,375,700  

Los datos muestran que **la Tienda 4 presenta el menor nivel de ingresos dentro de la cadena**, lo que indica un rendimiento comercial inferior en comparación con las demás tiendas.

La Tienda 1, por otro lado, lidera en ingresos, lo que sugiere una mayor capacidad de generación de ventas.

---

## Análisis de categorías de productos

En todas las tiendas, la categoría **Muebles** aparece como la más vendida. Esto indica que este tipo de producto representa una parte importante del volumen de ventas dentro de la cadena.

Las categorías menos vendidas son:

- **Tienda 1:** Artículos para el hogar  
- **Tienda 2:** Artículos para el hogar  
- **Tienda 3:** Instrumentos musicales  
- **Tienda 4:** Instrumentos musicales  

Este patrón sugiere que algunas categorías tienen menor demanda relativa y podrían representar oportunidades de mejora en la estrategia de productos.

---

## Evaluación de satisfacción del cliente

La satisfacción de los clientes se evaluó mediante la calificación promedio registrada en cada tienda:

- **Tienda 1:** 3.98  
- **Tienda 2:** 4.04  
- **Tienda 3:** 4.05  
- **Tienda 4:** 4.00  

Las calificaciones se mantienen relativamente estables entre las tiendas, todas cercanas a **4 puntos sobre 5**, lo que indica que la experiencia del cliente es consistente y positiva en toda la cadena.

Aunque la Tienda 3 tiene la mejor calificación promedio, la diferencia entre tiendas es pequeña y no representa una variación significativa en términos de satisfacción del cliente.

---

## Productos más y menos vendidos

Los productos con mayor número de ventas en cada tienda son:

- **Tienda 1:** Microondas  
- **Tienda 2:** Iniciando en programación  
- **Tienda 3:** Kit de bancas  
- **Tienda 4:** Cama box  

Mientras que los productos con menor número de ventas son:

- **Tienda 1:** Auriculares con micrófono  
- **Tienda 2:** Juego de mesa  
- **Tienda 3:** Bloques de construcción  
- **Tienda 4:** Guitarra eléctrica  

Este análisis permite identificar patrones de consumo y entender qué productos generan mayor rotación dentro de cada tienda.

---

## Costo promedio de envío

El costo promedio de envío por tienda es el siguiente:

- **Tienda 1:** 26,018.61  
- **Tienda 2:** 25,216.24  
- **Tienda 3:** 24,805.68  
- **Tienda 4:** 23,459.46  

La **Tienda 4 presenta el costo promedio de envío más bajo**, lo que podría representar una mayor eficiencia logística. Sin embargo, esta ventaja no compensa su menor nivel de ingresos en comparación con las otras tiendas.

---

## Conclusión y recomendación

Después de analizar los diferentes indicadores de desempeño, se observa que **la Tienda 4 presenta el menor ingreso total entre las cuatro tiendas**, lo que la posiciona como la tienda con menor rendimiento comercial dentro de la cadena.

Aunque su costo de envío promedio es menor y sus calificaciones de clientes se mantienen dentro de un rango positivo, el nivel de facturación sigue siendo el factor más relevante para evaluar la rentabilidad del negocio.

Por lo tanto, **mi recomendación basada en el análisis de datos es que el Sr. Juan considere vender la Tienda 4**, ya que representa la opción con menor generación de ingresos dentro del conjunto de tiendas analizadas. Esto permitiría liberar recursos para invertir en nuevas oportunidades de negocio con mayor potencial de crecimiento.
